### Reading Files

```python
# now try plain file read
path = '/Users/jumar/Private/edu/DataScience/datasets/CR/prehled-o-celkovem-poctu-osvc-v-cr.csv'
# by default it opens in "read-only" mode ('r')
# we can also use 'w' (create a new file, overwrite existing!)
# and 'x' (create a new file but fail if the file already exists)
f = open(path)

# let's read the file line-by-line and stripe End-Of-Line characters
lines = [line.rstrip() for line in f]
for l in lines:
    print(l)
f.close

```

```python
# it's important to always close the file - we can do that more thoroughly using `with`
with open(path) as f:
    lines = [line.rstrip() for line in f]

# notice that we can still access lines outside of the `with` block ??
for l in lines:
    print(l);
    
```

```python
# reading files - read, seek, tell

# read text
with open(path) as osvc_file:
    print(osvc_file.read(10))
    
# open it in binary mode and read
# in this case it's the same but non-ASCII chars would be different
with open(path, 'rb') as osvc_file:
    print(osvc_file.read(10))
    # give my my current position
    print("You're here: {0}".format(osvc_file.tell()))
```

### Writing files

```python
# use write or writelines
with open('tmp.txt', 'w') as f:
    # writing non-blank lines of original file
    f.writelines(x for x in open(path) if len(x) > 1)
with open('tmp.txt') as f:
    lines = f.readlines();
lines
```

## Unicode and binary mode

```python
with open('tmp.txt', 'rb') as f:
    data = f.read(50)
    print(data.decode())
```

```python
with open('tmp_nonascii.txt', 'wt') as f:
    f.writelines('Sueña')

with open('tmp_nonascii.txt', 'rt') as f:
    print(f.read(5))
    f.seek(4)
    print(f.read(1))
    
```

```python
# read OSVC file, sort by date and show max
with open(path) as f:
    # https://docs.python.org/2/library/stdtypes.html#str.splitlines
    lines = f.read().splitlines()

def fields(line):
    return line.split(sep=',')

# How to make a dictionary out of file?
header = fields(lines[0])
print("Header: ", header)
data = []
for l in lines[1:]:    
    data_fields = dict(zip(header, fields(l)))
    data.append(data_fields)

for d in data:
    print(d)

# sort dictionary: https://stackoverflow.com/questions/613183/how-do-i-sort-a-dictionary-by-value
# https://www.geeksforgeeks.org/ways-sort-list-dictionaries-values-python-using-lambda-function/
sorted(data, key=lambda i: i['vykonavana_cinnost_hlavni'])
```
