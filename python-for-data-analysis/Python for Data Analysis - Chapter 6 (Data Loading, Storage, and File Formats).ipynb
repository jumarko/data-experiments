### Handling missing values

```python
# no handling - c at index 1 is empty, message at index 0 is 'NA'
pd.read_csv('examples/ex5.csv')
```

```python
# providing custom na_values
pd.read_csv('examples/ex5.csv', na_values=['12'])
```

```python
# or you can provide custom na_values per column
sentinels = {'c': ['3', '11'], 'message': ['foo']}
pd.read_csv('examples/ex5.csv', na_values=sentinels)
```
