## Bar charts

```python
# a bar chart is a good choice when you want to show
# how some quantity varies among discrete set of items

movies = ['Annie Halle', 'Ben-Hur', 'Casablanca', 'Gandhi', 'West Side Story']
num_oscars = [5, 11, 3, 8, 10]
plt.bar(range(len(movies)), num_oscars)
```

```python
# need to call plt.bar again otherwise it shows an empty graph
plt.bar(range(len(movies)), num_oscars)

plt.title('My Favorite Movies')
plt.ylabel('# of Academy Awards')
# label x-axis with movie names at bar centers
plt.xticks(range(len(movies)), movies)
# this time call show explicitly although it's not strictly need in a notebook
plt.show()
```

### Histograms

Barcharts can be a good choice for plotting histograms.
You can then explore how values are distributed.

Note: there's also https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# histogram = Counter(for grade in grades)

# bucket the values and prod
buckets = (min(grade // 10 * 10, 90) for grade in grades)
histogram = Counter(buckets)

plt.bar([x + 5 for x in histogram.keys()], # shift bars right by 5
        histogram.values(),
        10, # each bar has width of 10
        edgecolor=(0,0,0))
plt.axis ([-5, 105, 0, 5]) # x-axis from -5 to 105, y-axis from 0 to 5

plt.xticks([10 * i for i in range(11)]) # x labels are 0,10, ..., 10
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```

## Line charts

```python
variance = [2 ** x for x in range(9)]
print(variance)

bias_squared = list(reversed(variance))
print(bias_squared)

total_error = [x + y for x,y in zip(variance, bias_squared)]
print(total_error)

xs = [i for i, _ in enumerate(variance)]
print(xs)
```

```python
# draw multiple lines by calling .plot multiple times

import matplotlib.pyplot as plt

# adjust plot size to make sure it's large enough - figuresize gets values in inches
# https://www.geeksforgeeks.org/change-plot-size-in-matplotlib-python/
plt.figure(figsize=(13,10))

plt.plot(xs, variance, 'g-', label='variance') # geen solid line
plt.plot(xs, bias_squared, 'r-.', label='bias^s') # read dot-dashed line
plt.plot(xs, total_error, 'b:', label='total error') # blue dotted line

plt.legend(loc=9)
plt.xlabel('model complexity')
plt.xticks([])
plt.title('The Bias-Variance Tradeoff')
plt.show()
```

## Scatterplots


###  Simple scatterplot showing dependency between two variables

Here we draw relationship between number of friends and minutes spent on the website per day.

These variables are not directly comparable (different scales) so we don't care what matplotlib chooses for axis scale.

```python
import matplotlib.pyplot as plt
# number of friends that people have
friends = [70,65,72,63,71,64,60,64,67]
# number of minutes that people spent on the website
minutes = [175,170,205,120,220,130,105,145,190]
labels = ['a','b','c','d','e','f','g','h','i']

plt.scatter(friends,minutes)

# label each point
for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label, 
                xy=(friend_count, minute_count),
                xytext=(5,-5),
                textcoords='offset points')
    
plt.title('Daily minutes vs Number of Friends')
plt.xlabel('# of friends')
plt.ylabel('daily minutes spent on the site')
plt.show()
```

###  Scatterplot with comparable variables

Here, both axis are grades, so we don't let matplotlib choose the scale because it would be misleading


```python
import matplotlib.pyplot as plt

test_1_grades = [99,90,85,97,80]
test_2_grades = [100,85,60,90,70]
```

```python
# let pyplot choose the scale => misleading visualization

plt.scatter(test_1_grades, test_2_grades)
plt.title('Axes are not comparable')
plt.xlabel('test 1 grade')
plt.ylabel('test 2 grade')

plt.show()
```

```python
# now we choose the scale

plt.scatter(test_1_grades, test_2_grades)
plt.title('Axes are not comparable')
plt.xlabel('test 1 grade')
plt.ylabel('test 2 grade')

# choose the equals scale => it becomes apparent, that the most variation occures on 'test 2'
plt.axis('equal')

plt.show()
```
