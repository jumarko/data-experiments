## Bar charts

```python
# a bar chart is a good choice when you want to show
# how some quantity varies among discrete set of items

movies = ['Annie Halle', 'Ben-Hur', 'Casablanca', 'Gandhi', 'West Side Story']
num_oscars = [5, 11, 3, 8, 10]
plt.bar(range(len(movies)), num_oscars)
```

```python
# need to call plt.bar again otherwise it shows an empty graph
plt.bar(range(len(movies)), num_oscars)

plt.title('My Favorite Movies')
plt.ylabel('# of Academy Awards')
# label x-axis with movie names at bar centers
plt.xticks(range(len(movies)), movies)
# this time call show explicitly although it's not strictly need in a notebook
plt.show()
```

### Histograms

Barcharts can be a good choice for plotting histograms.
You can then explore how values are distributed.

Note: there's also https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

# histogram = Counter(for grade in grades)

# bucket the values and prod
buckets = (min(grade // 10 * 10, 90) for grade in grades)
histogram = Counter(buckets)

plt.bar([x + 5 for x in histogram.keys()], # shift bars right by 5
        histogram.values(),
        10, # each bar has width of 10
        edgecolor=(0,0,0))
plt.axis ([-5, 105, 0, 5]) # x-axis from -5 to 105, y-axis from 0 to 5

plt.xticks([10 * i for i in range(11)]) # x labels are 0,10, ..., 10
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distribution of Exam 1 Grades")
plt.show()
```
